```python
#le o arquivo
import pandas as pd
dataset = pd.read_csv('callcenter_marketing.csv')
```

```python
#mostra as 5 primeiras linhas
dataset.head()

#mostra a quantidade de registros nulos por coluna (so se trata nulo em categorico, nao numerico)
#dataset.isnull().sum()
```

```python
#separando registro numerico de categorico
dataset_numerico = dataset[['idade','duracao','qtd_contatos_campanha','dias_ultimo_contato','qtd_contatos_total','indice_precos_consumidor','indice_confianca_consumidor','euribor3m']]
dataset_categorico = dataset[['profissao','estado_civil','educacao','inadimplente','emprestimo_moradia','emprestimo_pessoal','meio_contato','mes','dia_da_semana','campanha_anterior']]
```

```python
#dataset_numerico.head()
#dataset_categorico.head()
```

```python
#tratar dados nulos
for col_name in dataset_categorico.columns:
    dataset_categorico[col_name].fillna('NAO INFORMADO', inplace=True)
```

```python
#só pode ser usado o get dummies com pelo menos 3 valores distintos
#get dummies - nao se pode avaliar string (o comando transforma cada variação da string em uma coluna e atribui 0/1)
dataset_categorico_corrigido = pd.get_dummies(dataset_categorico)
dataset_categorico_corrigido.head()
```

```python
#juntar dados de treinamento
X = pd.concat([dataset_categorico_corrigido, dataset_numerico], axis=1)
X.head()
```

```python
def retorna_num(value):
    if value == 'sim': 
        return 1
    return 0
```

```python
from sklearn.model_selection import train_test_split

y_string = dataset.iloc[:,[18]]
y_string['resultado'] = y_string['resultado'].apply(retorna_num)

y = y_string.iloc[:,[0]]
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=0)
```

```python
#zera os indices
X_test.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression() #cria o modelo
lr.fit(X_train,y_train) #faz o fit do modelo(reta)

previsto = lr.predict(X_test) #predicao nos dados teste
```

```python
def categorize(value):
    if value <= 0.50: 
        return 0 # not admitted
    return 1 # admitted
```

```python
import pandas as pd
import numpy as np

serie_prevista = pd.Series((i[0] for i in previsto)) 
serie_prevista = serie_prevista.apply(categorize)

# cria nova coluna no y_test
y_test['teste'] = (y_test['resultado']).apply(categorize)


print('Erros: '+str(np.count_nonzero(serie_prevista.values!=y_test['teste'])))
print('Acertos: '+str(np.count_nonzero(serie_prevista.values==y_test['teste'])))
print('Total: '+str(serie_prevista.size))
print('O erro percentual é: '+str((np.count_nonzero(serie_prevista.values!=y_test['teste'])/serie_prevista.size)*100))
print('A acurácia é: '+str((np.count_nonzero(serie_prevista.values==y_test['teste'])/serie_prevista.size)*100))
```

```python
#regressao logistica
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(multi_class='multinomial',solver='lbfgs')
logreg.fit(X_train, y_train)
acuracia_logistica = logreg.score(X_test, y_test)
display(logreg.score(X_test, y_test))
```
